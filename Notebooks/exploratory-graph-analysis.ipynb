Handling Outliers using Interquartile Range


```python
p0=layover.min()
p100=layover.max()
q1=layover.quantile(0.25)
q2=layover.quantile(0.5)
q3=layover.quantile(0.75)
iqr=q3-q1
```

```python
lc = q1 - 1.5*iqr #lower cutoff
uc = q3 + 1.5*iqr #uppr cutoff
```
